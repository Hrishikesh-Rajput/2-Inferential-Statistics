Suppose the IQ in a certain population is normally distributed with the mean of 100 and standard deviation oof 15

A researcher wants to know if a new drug affects IQ levels, so he recruits 20 patients to try it and records their IQ levels.

The following code shows how to perform a one sample Z test in python to determine if the new drug causes a significant difference in IQ levels.

# ztest

```python
from statsmodels.stats.weightstats import ztest as ztest
# enter IQ levels for 20 patients
data=[88,92,94,94,96,97,97,97,99,99,105,109,109,109,110,112,112,113,114,115]
```

```python
value=ztest(data,value=100)
```

```python
# Example 2
# Ages of the College student (Population)
# 1 class student mean of all the ages
```

```python
import pandas as pd
import math
import scipy.stats as stats 
np.random.seed(6)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
classA_ages
```

```python
classA_ages.mean()
```

```python
_,p_value=ttest_1samp(classA_ages,school_ages.mean()) # popmean = population mean 
```

```python
school_ages.mean()
```

```python
if p_value<=0.05:
    print("Reject the null hypothesis / H0")
else:
    print("Accept null hyposthesis")
```
